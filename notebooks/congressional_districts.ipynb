```python
# Create a dictionary to store the state names and their corresponding list lengths
state_lengths = {}

# Iterate through the congressional_districts dictionary
for state, districts in congressional_districts.items():
    # Calculate the length of the list associated with each state
    length = len(districts)
    # Store the state and its corresponding list length in the state_lengths dictionary
    state_lengths[state] = length

# Print the state names and their corresponding list lengths
for state, length in state_lengths.items():
    print(f"State: {state}, Number of Districts: {length}")
```
